```python
from gensim.models import KeyedVectors, Word2Vec
import gensim.downloader as api
from gensim.utils import tokenize
from gensim.parsing.preprocessing import remove_stopwords
from gensim.parsing.preprocessing import STOPWORDS
from gensim.models import WordEmbeddingSimilarityIndex
from gensim.similarities import SoftCosineSimilarity, SparseTermSimilarityMatrix, MatrixSimilarity, Similarity
from gensim.corpora import Dictionary
import json
import pandas as pd
```

```python
def text_preprocessing(texts):
    """
    process text into a [[str]], lower case, remove punctuation,
    remove word which only has one character
    remove 'N/A' string!
    stop words (preserve negative meaning)? > may not needed
    lemmatization? > may not needed
    :param text: [str], each str is a description text
    :return:
    """
    # return a list of generator does not work well with gensim's Dictionary class
    processed_texts = [list(tokenize(text.lower())) for text in texts]

    result = []
    for sentence in processed_texts:
        result.append([word for word in sentence if (len(word) > 1) and (word not in STOPWORDS)])

    return result
```

```python
def train_model_process_texts(path='/Users/mm17060/Projects/TVA/test/input_id_and_description.json'):
    """

    :param path: json file path
    :return: [[str]]
    """
    with open(path, 'r') as f:
        data = json.load(f)
    texts = []
    input_ids = {}
    for ind, item in enumerate(data['data']):
        texts.append(item["input_description"])
        input_ids[ind] = item["input_id"]
        
    # texts = [item["input_description"] for item in data['input_description']] # [str]
    preprossed_texts = text_preprocessing(texts)

    return input_ids, preprossed_texts
```
